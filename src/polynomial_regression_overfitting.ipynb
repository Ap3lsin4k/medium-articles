# Polynomial Regression Overfitting

```python
#imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.lines as lines
%matplotlib inline

import statsmodels.api as sm
```

```python
#Generate data
a = 50
#b = int(a/5)
"""x = [max(0,x) for x in np.linspace(0, 50, a)+ 5*np.random.randn(a)]+ [
    max(0,x) for x in np.linspace(0, 50, a)+ 5*np.random.randn(a)]+ [
    max(0,x) for x in np.linspace(0, 50, a)+ 5*np.random.randn(a)]+ [
    max(0,x) for x in np.linspace(0, 50, a)+ 5*np.random.randn(a)]+ [
    max(0,x) for x in np.linspace(0, 50, a)+ 5*np.random.randn(a)]+ [
    max(0,x) for x in np.linspace(0, 50, a)+ 5*np.random.randn(a)] """
x = [max(0,x) for x in np.linspace(0, 50, a)+ 5*np.random.randn(a)]+ [
    max(0,x) for x in np.linspace(0, 40, a)+ 5*np.random.randn(a)]+ [
    max(0,x) for x in np.linspace(0, 30, a)- abs(5*np.random.randn(a))]+ [
    max(0,x) for x in np.linspace(0, 20, a)- abs (5*np.random.randn(a))]+ [
    max(0,x) for x in np.linspace(0, 10, a)-  abs(5*np.random.randn(a))]+ [
    max(0,x) for x in np.linspace(0, 5, a)-abs (5*np.random.randn(a))] 

x = np.array(x)
noise = [max(10,n) for n in (60-x)*np.random.randn(a*6) ]

y = (3/16)*x**(2)-(15/2)*x+100 + abs((60-x)*np.random.randn(a*6))+ 50*np.random.randn(a*6)
y = [max(5,y+50) for y in y]

fig, ax = plt.subplots(nrows=1, ncols=1,figsize=(16,8))
sns.scatterplot(y = y,x = x)
```

```python
data = pd.DataFrame(data=x, columns=["x"])
data['y'] = y
```

```python
data.to_csv('../data/house_price.csv')
```

```python
data = pd.read_csv('../data/house_price.csv')
```

```python
#For simplicity we will only consider 2 variables: house age and number of convenience stores
x = data['x']
y = data['y']

#Relationship with explanatory variables
fig, ax = plt.subplots(nrows=1, ncols=1,figsize=(16,8))
sns.scatterplot(x,y)
```

```python
# y = ax+b

x = data['x']
x = sm.add_constant(x)
y = data['y']

#fit model 
model = sm.OLS(y,x).fit()

#plot data
fig, ax = plt.subplots( nrows=1, ncols=1,figsize=(16,8))
plt.ylim(-10, 310)
sns.scatterplot(x['x'],y)

#plot prediciton line
x_line = np.linspace(0, 60, 1000)
x_line = pd.DataFrame(data=x_line, columns=["x"])
x_line  = sm.add_constant(x_line)

y_line = model.predict(x_line)

plt.plot(x_line['x'],y_line, 'r-', lw=2)
```

```python
# y = ax^2+bx + c
x = data[['x']]
x['x2'] = x**2

x = sm.add_constant(x)
y = data['y']

#fit model 
model = sm.OLS(y,x).fit()

#plot data
fig, ax = plt.subplots( nrows=1, ncols=1,figsize=(16,8))
plt.ylim(-10, 310)
sns.scatterplot(x['x'],y)

#plot prediciton line
x_line = np.linspace(0, 50, 1000)
x_line = pd.DataFrame(data=x_line, columns=["x"])
x_line  = sm.add_constant(x_line)
x_line['x2'] = x_line['x']**2

y_line = model.predict(x_line)

plt.plot(x_line['x'],y_line, 'r-', lw=2)
```

```python
x = data[['x']]
x = sm.add_constant(x)
y = data['y']


x_line = np.linspace(0, 50, 1000)
x_line = pd.DataFrame(data=x_line, columns=["x"])
x_line  = sm.add_constant(x_line)


for i in range(1,21):
    if i != 1:
        x['x{}'.format(i)] = x['x']**i
        
        x_line['x{}'.format(i)] = x_line['x']**i

    #fit model 
    model = sm.OLS(y,x).fit()

    #get predictions 
    y_line = model.predict(x_line)
    
    
    fig, ax = plt.subplots( nrows=1, ncols=1,figsize=(16,8))
    plt.ylim(-10, 310)
    plt.title('Terms: {}'.format(i),size =15)
    sns.scatterplot(y=y,x=x['x'])

    plt.plot(x_line['x'],y_line, 'r-', lw=2)
    plt.savefig('../figures/poly/v2plot{}.png'.format(i),format='png', dpi=150)
```
